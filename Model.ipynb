```python
def get_names(objec):
    LIST = []
    for i in json.loads(objec):
        LIST.append(i['name'])
    return LIST 
#X ten genres,keywords .... isimleri aliyoruz
```

```python
df['genres'].apply(get_names)[0]
```

```python
df['genres'] = df['genres'].apply(get_names)
```

```python
df['keywords'] = df['keywords'].apply(get_names)
```

```python
def get_cast_names(objec):
    LIST = []
    i  = 0
    for j in json.loads(objec):
        LIST.append(j['name'])
        i +=1
        if i >3:
            break
    return LIST
#main charecter isimleri aliyoruz ilk 4 charecter        
```

```python
df['cast'].apply(get_cast_names)[1]
```

```python
def get_director(objec):
    LIST = []
    for i in json.loads(objec):
        if i['job'] == 'Director':
            LIST.append(i['name'])
            break
    return LIST
```

```python
df['crew'].apply(get_director)
```

```python
df['crew'] = df['crew'].apply(get_director)
```

```python
df['cast'] = df['cast'].apply(get_cast_names)
```

```python
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
df['genres'] = df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['keywords'] = df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df['overview'] = df['overview'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew'] = df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast'] = df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```
